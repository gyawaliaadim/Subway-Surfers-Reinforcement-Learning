```python
"""
End-screen detection of subway surfers using template matching.

Loads a reference image (`cropped_image.png`) and compares it against
a specific region of a game screenshot (`endscreen.png`). The selected
region is cropped, resized if necessary, and matched using OpenCV's
normalized correlation template matching. If the similarity score
exceeds the defined threshold, the script reports a successful match.
"""
```

```python

import cv2
import numpy as np
import os
 
```

```python

# BASE_DIR = os.path.dirname(os.path.abspath(__file__))
BASE_DIR =os.getcwd()
image_path = os.path.join(
    BASE_DIR,
    "..",
    "game_regions",
    "game_screen.jpg"
)
coin_region = {
    'top': 268,
    'left': 1672,
    'width': 178,
    'height': 34
}
image_path = os.path.abspath(image_path)
```

```python

# Crop region
y = coin_region['top']
x = coin_region['left']
h = coin_region['height']
w = coin_region['width']

# Direct image path (ONLY ONE IMAGE)

```

```python

# Load image
img = cv2.imread(image_path)

if img is None:
    print(f"Could not load image: {image_path}")
    exit()

print("Press any key to close.")

# Crop
cropped_img = img[y:y+h, x:x+w]

# Convert to grayscale
gray_cropped = cv2.cvtColor(cropped_img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Cropped Region", gray_cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
