```python
import cv2
import os

# Load the image containing the number
image = cv2.imread("number.png", cv2.IMREAD_GRAYSCALE)
BASE_DIR =os.getcwd()
numbers_path = os.path.join(
    BASE_DIR,
    "..",
    "game_regions",
    "numbers"
)
result = []

# Threshold to make matching more robust (optional)
_, image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

for digit in range(10):
    template = cv2.imread(f"{numbers_path}/{digit}.png", cv2.IMREAD_GRAYSCALE)
    _, template = cv2.threshold(template, 128, 255, cv2.THRESH_BINARY)

    h, w = template.shape

    matches = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)

    threshold = 0.99
    locations = zip(*((matches >= threshold).nonzero()[::-1]))

    for pt in locations:
        result.append((pt[0], str(digit)))   # x-position, digit

# Sort digits from left to right
result.sort(key=lambda x: x[0])

number = "".join(d for _, d in result)

print(number)
```
